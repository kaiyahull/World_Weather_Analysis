```python
# Import dependencies and API keys
import pandas as pd
import numpy as np
import requests
from config import weather_api_key
from citipy import citipy
```

```python
# Create a list of 2000 random latitudes and longitudes 
lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)
lats_lngs = zip(lats, lngs)
lats_lngs
```

```python
# Add the latitudes and longitudes to a list
coordinates = list(lats_lngs)
```

```python
# Create a list of cities by looping through the list of coordinates (lats, lngs) 
# to find the nearest city 
cities = []

for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    
    if city not in cities:
        cities.append(city)
```

```python
# Starter url for Open Weather 
base_url = "http://api.openweathermap.org/data/2.5/weather?" + "&appid=" + weather_api_key + "&units=" + "Imperial"
```
